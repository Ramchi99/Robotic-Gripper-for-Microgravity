```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#data = "/home/zrene/catkin_ws/src/franka_ft_reader/output/log_202403281154.csv"
#data = "/home/zrene/catkin_ws/src/franka_ft_reader/output/log_202404091519.csv"
#data = "/home/zrene/catkin_ws/src/franka_ft_reader/output/log_202404091526.csv"
#data = "/home/zrene/catkin_ws/src/franka_ft_reader/output/log_202404091537.csv"
#data = "/home/zrene/catkin_ws/src/franka_ft_reader/output/log_202404101538.csv"
data = "/home/zrene/catkin_ws/src/franka_ft_reader/output/log_202404101626.csv"

df = pd.read_csv(data)
```

```python
plt.plot(df.Fx, label = "Fx")
plt.plot(df.Fy, label = "Fy")
plt.plot(df.Fz, label = "Fz")
plt.plot(df.ref_Fx, label="Reference Fx")
plt.plot(df.ref_Fy, label="Reference Fy")
plt.plot(df.ref_Fz, label="Reference Fz")
plt.title("Force Over Time")
plt.xlabel("Time [steps]")
plt.ylabel("Forces [N]")
plt.legend()
plt.grid()
```

```python
plt.plot(df.x, label = "x")
plt.plot(df.y, label = "y")
plt.plot(df.z, label = "z")
plt.plot(df.ref_x, label="Ref_x")
plt.plot(df.ref_y, label="Ref_y")
plt.plot(df.ref_z, label="Ref_z")
plt.title("Position Over Time")
plt.xlabel("Time [steps]")
plt.ylabel("Position [m]")
plt.legend()
plt.grid()
```

```python
plt.subplot(3,1,1)
plt.plot(df.Fx, df.x, 'x')
plt.ylabel("Position [m]")
plt.xlabel("Force [N]")
plt.title("Force in X Direction Vs Position X")
plt.grid()
plt.tight_layout()

plt.subplot(3,1,2)
plt.plot(df.Fy, df.y, 'y')
plt.ylabel("Position [m]")
plt.xlabel("Force [N]")
plt.title("Force in Y Direction Vs Position Y")
plt.grid()
plt.tight_layout()

plt.subplot(3,1,3)
plt.plot(df.Fz, df.z, 'x')
plt.ylabel("Position [m]")
plt.xlabel("Force [N]")
plt.title("Force in Z Direction Vs Position Z")
plt.grid()
plt.tight_layout()
plt.gcf().set_size_inches(15,10)
```
