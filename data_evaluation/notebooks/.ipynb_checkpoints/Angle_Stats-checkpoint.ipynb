```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the directory containing the statistics files
directory = "C://Users//sramc//Documents//01_Studium//50_Bachelor's Thesis//07_Test//Gripper Test Microspines Angles//Stats"
```

```python
# Function to import and plot the data
def import_and_calculate_statistics():
    # Define the rock names
    rock_names = ['Asphalt / Tarmac', 'Limestone Tuff', 'Volcanic Lava Rock', 'Breccia', 'Slate / Shale Stone', 'Sedimentary Rock']  # Add more as needed
    
    # Initialize dictionary to store statistics for each degree
    statistics = {}
    
    # Loop over degrees
    for degree in [40, 50, 60, 70, 80, 90]:
        
        # Initialize lists to store mean Fz values for all datasets and maximum Fz values for all datasets
        mean_Fz_values_all_datasets = []
        max_Fz_values_all_datasets = []
        
        # Loop over rock names
        for rock_name in rock_names:
            # Find all files for this degree and rock name
            file_pattern = f'C:\\Users\\sramc\\Documents\\01_Studium\\50_Bachelor\'s Thesis\\07_Test\\Gripper Test Microspines Angles\\{degree}\\{rock_name}_{degree}_*.csv'
            files = glob.glob(file_pattern)
            print(f"Searching for files with pattern: {file_pattern}")
            
            # Check if files are found
            if not files:
                print(f"No files found for {rock_name} at {degree} degrees.")
                continue
            
            # Loop over each file
            for file in files:
                # Read the CSV file into a DataFrame
                df = pd.read_csv(file)
                
                # Calculate mean Fz value and append to the list
                mean_Fz_values_all_datasets.append(df['Fz'].mean())
                
                # Calculate maximum Fz value and append to the list
                max_Fz_values_all_datasets.append(df['Fz'].max())
        
        # Calculate average and standard deviation of mean Fz values for all datasets
        avg_mean_Fz = sum(mean_Fz_values_all_datasets) / len(mean_Fz_values_all_datasets)
        std_dev_mean_Fz = pd.Series(mean_Fz_values_all_datasets).std()
        
        # Calculate average and standard deviation of maximum Fz values for all datasets
        avg_max_Fz = sum(max_Fz_values_all_datasets) / len(max_Fz_values_all_datasets)
        std_dev_max_Fz = pd.Series(max_Fz_values_all_datasets).std()

        # Calculate standard error for mean Fz
        std_error_mean_Fz = std_dev_mean_Fz / (len(mean_Fz_values_all_datasets) ** 0.5)
        
        # Calculate standard error for max Fz
        std_error_max_Fz = std_dev_max_Fz / (len(max_Fz_values_all_datasets) ** 0.5)
        
        statistics[degree] = {
            'avg_mean_Fz': avg_mean_Fz, 
            'std_dev_mean_Fz': std_dev_mean_Fz, 
            'std_error_mean_Fz': std_error_mean_Fz,
            'avg_max_Fz': avg_max_Fz, 
            'std_dev_max_Fz': std_dev_max_Fz, 
            'std_error_max_Fz': std_error_max_Fz
        }
        
        # Store statistics in the dictionary
        #statistics[degree] = {'avg_mean_Fz': avg_mean_Fz, 'std_dev_mean_Fz': std_dev_mean_Fz, 'avg_max_Fz': avg_max_Fz, 'std_dev_max_Fz': std_dev_max_Fz}
    
    return statistics
```

```python
def plot_statistics(statistics):
    degrees = list(statistics.keys())
    avg_mean_Fz = [stats['avg_mean_Fz'] for stats in statistics.values()]
    std_dev_mean_Fz = [stats['std_dev_mean_Fz'] for stats in statistics.values()]
    std_error_mean_Fz = [stats['std_error_mean_Fz'] for stats in statistics.values()]
    avg_max_Fz = [stats['avg_max_Fz'] for stats in statistics.values()]
    std_dev_max_Fz = [stats['std_dev_max_Fz'] for stats in statistics.values()]
    std_error_max_Fz = [stats['std_error_max_Fz'] for stats in statistics.values()]

    print("Average Avg Fz:")
    for degree, stats in statistics.items():
        print(f"Angle: {degree}")
        print(f"Average Avg Fz: {stats['avg_mean_Fz']:.2f} +/- {stats['std_dev_mean_Fz']:.2f} (std dev) +/- {stats['std_error_mean_Fz']:.2f} (std error)")
        print()
    
    # Plot for average mean Fz
    plt.errorbar(degrees, avg_mean_Fz, yerr=std_dev_mean_Fz, fmt='o-', label='Average Avg Fz (Std Dev)', color='blue')
    plt.errorbar(degrees, avg_mean_Fz, yerr=std_error_mean_Fz, fmt='o--', label='Average Avg Fz (Std Error)', color='orange')
    plt.title('Average Avg Fz Statistics for Each Angle')
    plt.xlabel('Angle (Degrees)')
    plt.ylabel('Average Avg Fz')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print("Average Max Fz:")
    for degree, stats in statistics.items():
        print(f"Angle: {degree}")
        print(f"Average Max Fz: {stats['avg_max_Fz']:.2f} +/- {stats['std_dev_max_Fz']:.2f} (std dev) +/- {stats['std_error_max_Fz']:.2f} (std error)")
        print()

    # Plot for average max Fz
    plt.errorbar(degrees, avg_max_Fz, yerr=std_dev_max_Fz, fmt='o-', label='Average Maximum Fz (Std Dev)', color='green')
    plt.errorbar(degrees, avg_max_Fz, yerr=std_error_max_Fz, fmt='o--', label='Average Maximum Fz (Std Error)', color='red')
    plt.title('Average Maximum Fz Statistics for Each Angle')
    plt.xlabel('Angle (Degrees)')
    plt.ylabel('Average Maximum Fz')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_average_max_fz_for_each_rock(directory):
    # Get all the statistics files in the directory
    statistics_files = os.listdir(directory)
    
    # Initialize a dictionary to store average max Fz values and standard deviations for each rock
    rock_stats = {}
    
    # Iterate over each statistics file
    for filename in statistics_files:
        # Extract rock name and angle from the filename
        rock, angle = filename.split('_')[:2]
        
        # Read the statistics file
        statistics_df = pd.read_csv(os.path.join(directory, filename))
        
        # Extract Average Max Fz and Std Max Fz for the current rock and angle
        avg_max_fz = statistics_df.loc[statistics_df['Statistic'] == 'Average Max Fz', 'Value'].iloc[0]
        std_max_fz = statistics_df.loc[statistics_df['Statistic'] == 'Std Max Fz', 'Value'].iloc[0]
        
        # Add the values to the dictionary for the corresponding rock and angle
        if rock not in rock_stats:
            rock_stats[rock] = {'angles': [], 'avg_max_fz': [], 'std_max_fz': []}
        rock_stats[rock]['angles'].append(int(angle))
        rock_stats[rock]['avg_max_fz'].append(avg_max_fz)
        rock_stats[rock]['std_max_fz'].append(std_max_fz)
    
    # Plot average max Fz for each rock
    for rock, data in rock_stats.items():
        plt.errorbar(data['angles'], data['avg_max_fz'], yerr=data['std_max_fz'], fmt='o-', label=f'{rock} Rock')

    plt.title('Average Max Fz for Each Rock')
    plt.xlabel('Angle')
    plt.ylabel('Average Max Fz')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def calculate_weighted_average_max_fz(directory):
    # Get all the statistics files in the directory
    statistics_files = os.listdir(directory)
    
    # Initialize dictionaries to store 'Average Max Fz' and their standard deviations for each angle
    angle_data = {}
    
    # Iterate over each statistics file
    for filename in statistics_files:
        # Extract angle from the filename
        angle = filename.split('_')[1]
        
        # Read the statistics file
        statistics_df = pd.read_csv(os.path.join(directory, filename))
        
        # Extract 'Average Max Fz' and 'Std Max Fz'
        avg_max_fz = statistics_df.loc[statistics_df['Statistic'] == 'Average Max Fz', 'Value'].iloc[0]
        std_max_fz = statistics_df.loc[statistics_df['Statistic'] == 'Std Max Fz', 'Value'].iloc[0]
        
        # Add data to the dictionary for the corresponding angle
        if angle not in angle_data:
            angle_data[angle] = {'avg_fz': [], 'std_fz': []}
        angle_data[angle]['avg_fz'].append(avg_max_fz)
        angle_data[angle]['std_fz'].append(std_max_fz)
    
    # Initialize lists to store weighted averages and their standard errors
    weighted_avg_max_fz = []
    std_error_weighted_avg_max_fz = []
    
    # Iterate over each angle
    for angle, data in angle_data.items():
        # Calculate weights as the inverse of the squared standard deviations
        weights = 1 / np.square(data['std_fz'])
        
        # Calculate weighted average
        weighted_avg = np.average(data['avg_fz'], weights=weights)
        
        # Estimate standard error of the weighted average using error propagation formulas !!!!!!!!!!!!!!!!!!!!
        #std_error_weighted_avg = np.sqrt(np.sum(np.square(weights * data['std_fz'])) / np.square(np.sum(weights)))
        # Calculate standard error of the weighted average
        std_error_weighted_avg = np.sqrt(1 / np.sum(weights))
        
        # Append the results to the lists
        weighted_avg_max_fz.append(weighted_avg)
        std_error_weighted_avg_max_fz.append(std_error_weighted_avg)

        # Print the results for each angle
        print(f"Angle: {angle}")
        print(f"Weighted Average Max Fz: {weighted_avg:.2f} +/- {std_error_weighted_avg:.2f}")
        print()
    
    return angle_data, weighted_avg_max_fz, std_error_weighted_avg_max_fz

def plot_weighted_average_max_fz(angle_data, weighted_avg_max_fz, std_error_weighted_avg_max_fz):
    angles = list(angle_data.keys())
    
    plt.errorbar(angles, weighted_avg_max_fz, yerr=std_error_weighted_avg_max_fz, fmt='o-', label='Weighted Average Max Fz')

    plt.title('Weighted Average Max Fz for Each Angle')
    plt.xlabel('Angle')
    plt.ylabel('Weighted Average Max Fz')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
